```python
import pandas as pd

df=pd.read_csv("kc_house_data.csv")
df.info()
```

```python
df.drop(['id','date'],axis=1,inplace=True)
df.isnull().sum()
```

```python
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar=MinMaxScaler()
```

```python
cols=['sqft_living','sqft_lot','grade','sqft_above','sqft_basement','yr_built','lat','long','sqft_living15','sqft_lot15']
df[cols]=scalar.fit_transform(df[cols])
df.head()
```

```python
x=df.drop(['price','zipcode','yr_renovated'],axis=1)
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
df.shape
```

```python
from tensorflow import keras

model=keras.Sequential([keras.layers.Dense(100,activation='relu',input_shape=(16,)),
                        keras.layers.Dense(100,activation='relu'),
                        keras.layers.Dense(100,activation='relu'),
                      keras.layers.Dense(1,activation='linear')])
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['R2Score'])
```

```python
model.fit(x_train,y_train,epochs=100)
```

```python
model.evaluate(x_test,y_test)
```

```python
from sklearn.metrics import r2_score
y_pred=model.predict(x_test)
r2_score(y_test,y_pred)*100
```
